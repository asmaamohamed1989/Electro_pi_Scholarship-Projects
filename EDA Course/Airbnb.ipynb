This project involves performing an exploratory data analysis (EDA) on Airbnb listings data for a particular city. The analysis will focus on factors such as price, availability, location, and property type to identify trends and patterns in the demand for Airbnb listings in the city. The project will include data cleaning, visualization, and statistical analysis using tools such as Python and Jupyter Notebook. Data: http://insideairbnb.com/get-the-data/

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Data Collection

df = pd.read_csv('/content/listings.csv')
```

```python
# Step 2: Data Cleaning
# Handle missing values and data type conversions
df['price'] = df['price'].str.replace('$', '').str.replace(',', '').astype(float)
df['availability_365'] = pd.to_numeric(df['availability_365'])
```

```python
# Step 3: Exploratory Data Analysis
# Summary statistics
summary_stats = df.describe()
summary_stats
```

```python
#function to calculate value count for each columns


def calculate_value_counts(data):
    value_counts = {}
    for column in data.columns:
        value_counts[column] = data[column].value_counts()
    return value_counts

value_counts = calculate_value_counts(df)
for column, counts in value_counts.items():
    print(f"Value counts for {column}:\n{counts}\n")


```

```python
# Visualizations
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Distribution of Listing Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of price vs availability
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='availability_365', y='price')
plt.title('Price vs Availability')
plt.xlabel('Availability (in days)')
plt.ylabel('Price')
plt.show()
```

```python
# Price Analysis
price_by_property_type = df.groupby('property_type')['price'].median().sort_values(ascending=False)
```

```python
# Property Type and Room Type Analysis
plt.figure(figsize=(30, 30))
sns.boxplot(data=df, x='property_type', y='price')
plt.title('Price Distribution by Property Type')
plt.xticks(rotation=90)
plt.xlabel('Property Type')
plt.ylabel('Price')
plt.show()

```

```python
plt.figure(figsize=(20, 20))
sns.countplot(data=df, x='room_type', palette='Set2')
plt.title('Distribution of Room Types')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation Analysis
correlation_matrix = df.corr()

plt.figure(figsize=(30, 30))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Conclusion and Insights
# Summarize key findings and insights from the analysis
print("Summary of Key Findings:")
print(summary_stats)
print("\nMedian Price by Property Type:")
print(price_by_property_type)
print("\nAverage Availability by Month:")

```

```python
# Analyzing Reviews
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='number_of_reviews', y='price')
plt.title('Price vs Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.show()
```

```python
# Analyzing Amenities
amenities = df['amenities'].str.replace('[{}"]', '').str.split(',')
amenities_list = [amenity.strip() for row in amenities for amenity in row]
unique_amenities = set(amenities_list)

amenity_counts = {}
for row in amenities:
    for amenity in row:
        amenity = amenity.strip()
        if amenity in amenity_counts:
            amenity_counts[amenity] += 1
        else:
            amenity_counts[amenity] = 1

top_amenities = sorted(amenity_counts.items(), key=lambda x: x[1], reverse=True)[:10]
top_amenities_labels = [item[0] for item in top_amenities]
top_amenities_values = [item[1] for item in top_amenities]

plt.figure(figsize=(12, 6))
sns.barplot(x=top_amenities_values, y=top_amenities_labels, palette='viridis')
plt.title('Top 10 Amenities')
plt.xlabel('Frequency')
plt.ylabel('Amenity')
plt.show()
```

```python
# Analyzing Host-Related Features
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='host_response_time', y='price', palette='Set2')
plt.title('Price Distribution by Host Response Time')
plt.xticks(rotation=45)
plt.xlabel('Host Response Time')
plt.ylabel('Price')
plt.show()
```

```python
# Exploring Host Characteristics
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='host_identity_verified', y='price', palette='Set2')
plt.title('Price Distribution by Host Identity Verified')
plt.xlabel('Host Identity Verified')
plt.ylabel('Price')
plt.show()

```

```python
# Property Characteristics
property_type_counts = df['property_type'].value_counts()

plt.figure(figsize=(30, 30))
sns.barplot(x=property_type_counts.index, y=property_type_counts.values, palette='viridis')
plt.title('Property Type Distribution')
plt.xlabel('Property Type')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()

```
